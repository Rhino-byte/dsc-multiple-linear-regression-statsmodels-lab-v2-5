## The Ames Housing Dataset

The [Ames Housing dataset](http://jse.amstat.org/v19n3/decock.pdf) is a newer (2011) replacement for the classic Boston Housing dataset. Each record represents a residential property sale in Ames, Iowa. It contains many different potential predictors and the target variable is `SalePrice`.

````python
import pandas as pd
ames = pd.read_csv("ames.csv", index_col=0)
ames
````

````python
ames.describe()
````

We will focus specifically on a subset of the overall dataset. These features are:

```
LotArea: Lot size in square feet

1stFlrSF: First Floor square feet

GrLivArea: Above grade (ground) living area square feet
```

````python
ames_subset = ames[['LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].copy()
ames_subset
````

## Step 1: Visualize Relationships Between Features and Target

For each feature in the subset, create a scatter plot that shows the feature on the x-axis and `SalePrice` on the y-axis.

````python
# Your code here - import relevant library, create scatter plots
import matplotlib.pyplot as plt
fig,ax =plt.subplots(1,3,figsize=(15,10))

ax[0].scatter(x=ames_subset['LotArea'],y=ames_subset['SalePrice'])
ax[0].set_xlabel('LotArea')
ax[1].scatter(x=ames_subset['1stFlrSF'],y=ames_subset['SalePrice'])
ax[1].set_xlabel('1stFlrSF')
ax[2].scatter(x=ames_subset['GrLivArea'],y=ames_subset['SalePrice']) 
ax[2].set_xlabel('GrLivArea')
fig.tight_layout()
;
````

````python
# Your written answer here - do these seem like good candidates for linear regression?

""" 
There is a positive linear relationship between the GrLivArea and 1stFlrSF with the response variable sales price
but for the case of the LotArea most of the values  are close to zero
"""
````

## Step 2: Build a Simple Linear Regression Model

Set the dependent variable (`y`) to be the `SalePrice`, then choose one of the features shown in the subset above to be the baseline independent variable (`X`).

Build a linear regression using StatsModels, describe the overall model performance, and interpret its coefficients.

````python
# Get the correlation between the response variables and the response variable
ames_subset.corr()['SalePrice']
````

````python
# Your code here - define y and baseline X
y= ames_subset['SalePrice']
baseline_x =ames_subset['GrLivArea']
````

````python
# Your code here - import StatsModels, fit baseline model, display results
import statsmodels.api as sm 
baseline_model = sm.OLS(y,sm.add_constant(baseline_x))
baseline_results = baseline_model.fit()
print(baseline_results.summary())
````

````python
# Your written answer here - interpret model results
""" 
Based on the F-statistics p-value we can conclude that model is statistically significant 
The model explains about 50% of the variance in the sales price
The model coefficients (const and GrLivArea) are both statistically significant with t-statistics p-value below 0.05

For each increase of 1 square feet in the Area of Above ground we can see an associated increase the Sales price of
about 107.13
"""
````

## Step 3: Build a Multiple Linear Regression Model

For this model, use **all of** the features in `ames_subset`.

````python
# Your code here - define X
X = ames_subset.drop(columns='SalePrice')
````

````python
# Your code here - fit model and display results
Final_model = sm.OLS(y,sm.add_constant(X))
Final_results =Final_model.fit()
print(Final_results.summary())
````

````python
# Your written answer here - interpret model results. Does this model seem better than the previous one?

""" 
The overall performance of the model based on F-statistics p-value is statistically significant with a p-value
less than 0.05

For the coefficients (const , 1stFlrSF, GrLivArea) their t-statistics p-value are less than 0.05 but for the case 
of the LotArea th p-value is slightly above the alpha level indicating that its coefficient is not statistically significant

Generally the performance of the model does not improve much as the model with all the features can only explain 56%
of the total variance of the Sales price a difference of 6% from the baseline model
"""
````

## Step 4: Create Partial Regression Plots for Features

Using your model from Step 3, visualize each of the features using partial regression plots.

````python
# Your code here - create partial regression plots for each predictor
fig =plt.figure(figsize=(15,8))
sm.graphics.plot_partregress_grid(
    Final_results,
    exog_idx=list(X.columns.values),
    grid=(2,2),
    fig =fig
)
plt.show()
````

````python
# Your written answer here - explain what you see, and how this relates
# to what you saw in Step 1. What do you notice?

""" 
From the partial regression plot for the LotArea it has a zero slope implying it does not have beneficial impact to the model
as compared to 1stFlrSF and GrLivArea which have non-zero slope indicating that they are beneficial to our model
This is supported by the summary from the model.
"""
````
